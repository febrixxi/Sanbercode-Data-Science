```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(17,6))
ax = ax.flatten()
ax[0].set_title('Grafik tanx')
ax[0].plot(x, tan, color='blue', label='tanX')
ax[1].set_title('Grafik sinx')
ax[1].plot(x, sin, color='red', label='sinX')
ax[2].set_title('Grafik tanx')
ax[2].plot(x, cos, color='green', label='cosX')
fig.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05))
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(13,8))
ax.plot(x, y, color='blue')
ax.set_title('Grafik CosX')

ins2 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins2.plot(x, y, color='blue')
ins2.plot(x, y2, color='red')
ins2.set_title('Grafik CosX^2')

ins3 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins3.plot(x, y, color='blue')
ins3.plot(x, y3, color='green')
ins3.set_title('Grafik CosX^3')

ins4 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins4.plot(x, y, color='blue')
ins4.plot(x, y4, color='black')
ins4.set_title('Grafik CosX^4')

ins5 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins5.plot(x, y, color='blue')
ins5.plot(x, y5, color='yellow')
ins5.set_title('Grafik CosX^5')

plt.show()
```
