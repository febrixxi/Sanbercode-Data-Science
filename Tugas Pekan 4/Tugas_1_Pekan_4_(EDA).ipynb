```python
# code here
df = pd.read_csv('austin_weather.csv', na_values=['na','-'])
d = {
    'DewPointAvgF' : 'float',
    'HumidityAvgPercent' : 'float',
    'WindAvgMPH' : 'float'
}
df.astype(d).info()
df.sample(1000)
```

```python
#tanpa handling missing values
plt.figure(figsize=(10, 8))
plt.scatter(df['HumidityAvgPercent'], df['DewPointAvgF'], c=df['TempAvgF'], s=df['WindAvgMPH'].values*20, cmap='coolwarm')
plt.title('Austin Weather', fontsize=12)
plt.xlabel('Humidity Average (%)', fontsize=10)
plt.ylabel('Dew Point Average (F)', fontsize=10)
plt.xticks(fontsize=8)

#dengan handling missing values
df1 = df.fillna(method='ffill')

plt.figure(figsize=(10, 8))
plt.scatter(df1['HumidityAvgPercent'], df1['DewPointAvgF'], c=df1['TempAvgF'], s=df1['WindAvgMPH'].values*20, cmap='coolwarm')
plt.title('Austin Weather', fontsize=12)
plt.xlabel('Humidity Average (%)', fontsize=10)
plt.ylabel('Dew Point Average (F)', fontsize=10)
plt.xticks(fontsize=8)

plt.show()
```

Analisa: Saat menambahkan na_values=['na', '-'], kolom yang memiliki nilai 'na' dan '-' akan berubah menjadi NaN, kemudian dengan mengisi missing value dengan method ffilna, kita dapat melihat nilai yang sudah terisi melalui visualisasi scatterplot diatas

```python
#Code here
import seaborn as sns

#data dengan outliers
sns.set({'figure.figsize':(10,8)})
sns.set_theme(style="white")
sns.boxplot(x=df2['House_Price'])

plt.show()
```

```python
#handling outliers
Q1 = df2['House_Price'].quantile(0.25)
Q3 = df2['House_Price'].quantile(0.75)
IQR = Q3 - Q1
Lb=Q1-(IQR*1.5)
Ub=Q3+(IQR*1.5)

df_no_outliers = df2[(df2['House_Price'] > Lb) & (df2['House_Price'] < Ub)]

print('Batas bawah & batas atas\n')
print(Lb, Ub)

#data tanpa outliers
sns.set({'figure.figsize':(10,8)})
sns.set_theme(style="white")
sns.boxplot(data=df_no_outliers, x='House_Price')

plt.show()
```
